```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
```

```python
cg = CoinGeckoAPI()
```

```python
bitcoin_data = cg.get_coin_market_chart_by_id(id = 'bitcoin', 
                                              vs_currency = 'usd', 
                                              days = 30)
```

```python
bitcoin_data['prices']
```

```python
data = pd.DataFrame(bitcoin_data['prices'], columns = ['Timestamp','Price'])
```

```python
data['Date'] = pd.to_datetime(data['Timestamp'], unit= 'ms')
data
```

```python
candelstick_data = data.groupby(data.Date.dt.date).agg({'Price':['min','max','first','last']})
```

```python
candelstick_data
```

```python
import plotly.graph_objects as go
import plotly
fig = go.Figure(data=[go.Candlestick(x= candelstick_data.index,
                                              open = candelstick_data['Price']['first'],
                                              high = candelstick_data['Price']['max'],
                                              low = candelstick_data['Price']['min'],
                                              close = candelstick_data['Price']['last'])])
fig.update_layout(xaxis_rangeslider_visible = False, xaxis_title = 'Date',
                 yaxis_title = 'Price (USD $)', title = 'Bitcoin Candelstick Chart Over Past 30 Days')
```

```python
plotly.offline.plot(fig, filename= "./hahaha.html")
```
